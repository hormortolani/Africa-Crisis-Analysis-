# Step 3: Visualization Creation



## Research Question: How do exchange rates and currency crises relate to the occurrence of banking crises in african countries over the year

```python
df = df.sort_values(by='year')

plt.figure(figsize=(10, 6))

plt.plot(df['year'], df['exch_usd'], marker='o', label='Exchange Rates')
plt.plot(df['year'], df['banking_crisis'], marker='o', label='Banking Crises')
plt.plot(df['year'], df['currency_crises'], marker='o', label='Currency Crises')

plt.title('Trends of Exchange Rates, Banking Crises, and Currency Crises')
plt.xlabel('Year')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.show()
```

###  It appear that we have significantly less data prior to 1940. This is something we should keep in mind

### The line chart is an effective way to visualize the temporal trends and relationships between multiple variables, helping to gain a better understanding of the dataset in the context of the research question.

### I observe how each variable changes over time by examining the lines.

### Comparision between the three lines allows more insights into potential relationships or correlations between exchange rates, currency crises, and banking crises.

```python
banking_crises_by_country = df.groupby(['country', 'banking_crisis']).size().unstack().fillna(0)
banking_crises_by_country.plot(kind='bar', stacked=True, figsize=(10, 8))
plt.title('Banking_Crises Across African Countries')
plt.xlabel('Country')
plt.ylabel('Number of Crises Occurrences')
plt.legend(title='Banking_Crisis', loc='upper left')
plt.xticks(rotation=90, ha="right")
plt.show()
```

### The above bar chart displays the occurrences of banking crises across African countries which shows Central        African Republic as more bank crises compares to others
### Different colors represent the presence or absence of a banking crisis (blue color indicate crises while orange indicate no_crises)
### Each bar represents a country, and the height of the bar is the total number of banking crises. 


```python
plt.figure(figsize=(10, 6))
average_exchange_rate = df.groupby('country')['exch_usd'].mean().sort_values(ascending=False)
sns.barplot(x=average_exchange_rate.index, y=average_exchange_rate.values)
plt.title('Average Exchange Rate by Country')
plt.xlabel('Country')
plt.ylabel('Average Exchange Rate')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.tight_layout()
plt.show()
```

### This bar chart shows the average exchange rates for each African country.
### It provides a clear comparison of exchange rates across different countries.
### This visualization gives a clear view of the country with the highest average exchange rate

```python
plt.figure(figsize=(10, 6))
currency_crises_by_country = df.groupby(['country', 'year'])['currency_crises'].mean().reset_index()
sns.lineplot(x='year', y='currency_crises', hue='country', data=currency_crises_by_country, palette="YlGnBu")
sns.set(style="whitegrid")

plt.title('Currency Crises Across African Countries Over Time')
plt.xlabel('Year')
plt.ylabel('Mean Currency Crises')
plt.legend(title='Country', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

### Each line represents a country, allowing for a visual comparison of how currency crises evolve over time for each African country.

### By examining the line graph, we can observe a predominant occurrence of currency crises spanning from the year 1930 to 2018

```python
correlation_matrix = df[['exch_usd', 'currency_crises']].corr()
correlation_matrix
```

### The correlation matrix shows the pairwise correlation coefficients between these two variables(exch_usd and currency_crises)

### Despite being statistically significant, the small magnitude of the correlation suggests a weak association between exchange rates ('exch_usd') and the occurrence of currency crises ('currency_crises').

```python
plt.figure(figsize=(10, 6))

sns.pairplot(df[['exch_usd', 'currency_crises', 'banking_crisis', 'inflation_annual_cpi']], hue='banking_crisis', diag_kind='kde')
plt.suptitle('Pair Plot of Key Variables', y=1.02)
plt.show()
```

### The pair plot is particularly useful for identifying patterns and correlations among multiple variables.

```python
plt.figure(figsize=(16, 6))
sns.lineplot(x='year', y='exch_usd', data=df, hue='inflation_crises')
plt.title('Exchange Rates Over the Years, Grouped by Inflation Crises')
plt.xlabel('Year')
plt.ylabel('Exchange Rates')

plt.tight_layout()
plt.show()
```

### This line plot illustrates how exchange rates vary during periods of inflation crises and how exchange rates               change over time

```python
plt.figure(figsize=(10, 6))

country_inflation_crises = df.groupby('country')['inflation_crises'].sum().sort_values(ascending=False)

sns.boxplot(x=country_inflation_crises.values, y=country_inflation_crises.index, palette='viridis')
plt.title('Total Inflation Crises by Country')
plt.xlabel('Total Inflation Crises')
plt.ylabel('Country')

plt.show()
```

### Angolo, Zambia and Zimbabwe suffered more number of inflations compared to all other african countries.
### South Africa has the lowest number of inflation crisis.
